# Making a Map

```python
import pandas as pd
corona_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/06-28-2020.csv')
corona_df= corona_df.loc[corona_df['Country_Region']== 'India']
corona_df.dropna(subset=['Lat'], inplace = True)
import folium
#!pip install folium
m = folium.Map(location = [28.646519, 77.10898],
           tiles = 'Stamen toner',
           zoom_start = 8)
folium.Circle(location = [28.646519, 77.10898], radius = 10000, color = 'red', fill = True,
              popup = 'confirmed {}'.format(20)).add_to(m)
def circle_maker(x):
  folium.Circle(location = [x[0], x[1]],
                radius = float(x[2])*10,
                popup = '{}\nConfirmed Cases: {}'.format(x[3], x[2])).add_to(m)
corona_df[['Lat', 'Long_', 'Confirmed', 'Combined_Key']].apply(lambda x: circle_maker(x), axis = 1)
```

```python
def find_top_confirmed(n = 15):
  import pandas as pd
  corona_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/06-28-2020.csv')
  by_country = corona_df.groupby('Country_Region').sum()[['Confirmed', 'Deaths', 'Recovered', 'Active']]
  cdf = by_country.nlargest(n, 'Confirmed')[['Confirmed']]
  return cdf
```
